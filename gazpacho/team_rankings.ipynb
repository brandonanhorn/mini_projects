```python
date = 1
url = 'https://www.teamrankings.com/nba/stat/points-per-game?date=2015-11-' + str(date)
```

```python
def scrape_page(date=1):
    url = 'https://www.teamrankings.com/nba/stat/points-per-game?date=2015-11-' + str(date)
    html = get(url)
    soup = Soup(html)
    
    trs = soup.find('tr')
    tr_pop = trs.pop(0)
    tr = trs[0]
    
    def parse_tr(tr):
        return {
            'date': '2015-11-' + str(date),
            'rank': tr.find('td', {'class': 'rank text-center'}).text,
            'team': tr.find('td', {'class': 'text-left nowrap'}).text,
            'current_year_total': tr.find('td', {'class': 'text-right'})[0].text,
            'last_three': tr.find('td', {'class': 'text-right'})[1].text,
            'last_game': tr.find('td', {'class': 'text-right'})[2].text,
            'home_score_avg': tr.find('td', {'class': 'text-right'})[3].text,
            'away_score_avg': tr.find('td', {'class': 'text-right'})[4].text,
        }
    
    
    rankers = [parse_tr(tr) for tr in trs]
    return rankers
```

```python
first_11 = scrape_page(date=1)
```

```python
first_11
```

```python
from tqdm import tqdm
import time
```

```python
multiple_dates = []
for date in tqdm(range(1,31)):
    multiple_dates.extend(scrape_page(date=date))
    time.sleep(.6)
```
